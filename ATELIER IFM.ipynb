## Handling Missing Values
**Notre dataset souffre des valeurs manquantes.**
- Quelles sont les raisons ?
- Alors que faut-il faire ?
  - Nombreuses sont les raisons derrière cette problèmatique.
- **Solution ?**
  - La suppression de données sera la méthode en dernier recours car elle engendre une perte importante d'informations.
  - Plus nous avons d'informations, meilleure sera la qualité de prediction.
  - Neanmoins, dans certains cas on optera tout de même cette stratégie.
  
- On préfèra à la suppression de ces donnée, l'imputation de valeur.
  -  Cette méthode permet de réduire la perte d'informations.
- **Plusieurs méthodes existent** :
  - On peut utliser la mediane ou la moyenne pour les variables quantitatives (Attention aux ***outliers*** s'il s'agit de la moyenne),
  - Pour les variables qualitatives, on peut imputer par la valeur majoritaire en utilisant le mode.
  - Une méthode plus fine sera d'imputer en tenant en compte plusieurs paramètres. On pourra ainsi calculer la moyenne, la médiane, le mode, l'ecart type. Etablir un intervalle de confiance par groupes d'individu au sein de dataset pour affiner l'analyse.

```python
data_na =  africa.isnull().sum()
data_na = data_na/len(africa)* 100
data_na
```

```python
missdata= data_na.drop(data_na[data_na== 0].index).sort_values(ascending = False)
missdata =  pd.DataFrame( {'MissData_Ratio' : missdata })
print(missdata)
```

- **Missing values**(zmaloct, zraloct)

#### Imputation Zraloct, Zmaloct

```python
pmath_avg=africa['zmalocp'].mean()
pmath_std=africa['zmalocp'].std()
pmath_null_count=africa['zmalocp'].isnull().sum()


read_avg=africa['zraloct'].mean()
read_std=africa['zraloct'].std()
read_null_count=africa['zraloct'].isnull().sum()


math_avg=africa['zmaloct'].mean()
math_std=africa['zmaloct'].std()
math_null_count=africa['zmaloct'].isnull().sum()


print(pmath_null_count)
print(read_null_count)
print(math_null_count)
```

- Nous avons un intervalle à plus ou moins écart de la moyenne
- La loi uniforme se caractérise par une répartition uniforme des valeurs prises par une variable aléatoire continue sur un intervalle [a ;b]. Dès lors, sa fonction de densité est constante sur [a ;b] et nulle en dehors de cet intervalle puisque X n'y prend pas de valeurs.

```python
pmath_null_random_list=np.random.uniform(pmath_avg-pmath_std, pmath_avg+pmath_std , size=pmath_null_count)
read_null_random_list=np.random.uniform(read_avg-read_std, read_avg+read_std , size=read_null_count)
math_null_random_list=np.random.uniform(math_avg-math_std, math_avg+math_std , size=math_null_count)
```

```python
pmath_null_random_list
```

```python
print(pmath_avg)
print(read_std)
#pmath_avg-read_std
#pmath_avg+read_std
```

```python
africa.info()
```

```python
africa['zmalocp'][np.isnan(africa['zmalocp'])]=pmath_null_random_list
africa['zraloct'][np.isnan(africa['zraloct'])]=read_null_random_list
africa['zmaloct'][np.isnan(africa['zmaloct'])]=math_null_random_list 
```

Nous avons decidé de supprimer les valeurs manquantes ayant une faible proportion par rapport au total

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
```

## Kmeans

```python
X = africa[['lon', 'lat']]
```

```python
X.info()
```

```python
sc= StandardScaler()
X_scaled = pd.DataFrame(sc.fit_transform(X.values), columns=X.columns)
```

```python
X_scaled.head()
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X_scaled)
    distortions.append(kmeanModel.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(9,6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('WC55')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

We can observe that the “elbow” is the number 4 which is optimal for this case. Now we can run a K-Means using as n_clusters the number 4.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data = X, y="lat", x="lon")
plt.show()
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
k = KMeans(n_clusters = 6).fit(X_scaled)
clusters = k.predict(X_scaled)
```
